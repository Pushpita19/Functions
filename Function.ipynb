**Q.1 What is the difference between a function and a method in Python?**
- A function is an independent block of code that is not associated with any object whereas a method is a function that is associated with an object(class).

- A function is defined using the *def* keyword and can be called anywhere in the program and a method is defined inside a class and is called using the object of that *class*.

```python
#Example of function: -
def sum(a,b):
  return a+b
print(sum(5,3))
```

```python
#Example of method :-
class Calculator:
    def add(self, a, b):
        return a + b

calc = Calculator()
result = calc.add(5, 3)
print("Sum:", result)
```

- In python, function can be defined and called by different methods.

```python
#Defination
def greet(name):
    return f"Namaste {name}"
#calling
print(greet("Bharat"))
```

```python
def multiply(factor):
    def multiply(n):
        return n * factor
    return multiply

times3 = multiply(3)
print(times3(10))
```

```python
class Calculator:
    def __init__(self, factor):
        self.factor = factor

    def multiply(self, n):
        return n * self.factor

calc = Calculator(3)
print(calc.multiply(10))
```

**Q.6 Explain the concept of generators in Python and how they are defined.**
- A generator in python is a function that contains one or more yield statements. When called, it does not execute immediately but returns a generator object, which can be iterated using next() or a loop.
- It is memory efficient and uses simple sytax.
-  Its is defined by using "yield" inside a function and also by using generator expression.

```python
#using yield
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 2

gen = count_up_to(10)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#Using generator
gen = (x*2 for x in range(10))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

**Q.7 What are the advantages of using generators over regular functions?**
-Memory efficiency of generator is better than regular functions.
- Genertor is faster than regular functions.
- Generator is more efficient in the handling of large database.
- Generator make the code more concise and readable compared to using manual iterators.

```python
#Regular function
def square_numbers_reg(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result

print(square_numbers_reg(5))
```

```python
#Generator
def square_numbers_gen(n):
    for i in range(n):
        yield i ** 2

gen = square_numbers_gen(5)
print(next(gen))
print(next(gen))
print(next(gen))
```

**Q.8 What is a lambda function in Python and when is it typically used?**
- A lambda function in python is an anonymous function defined using the lambda keyword. It is used for short and simple operation.
- It is also used in passing short functions as arguments to map(), filter(), reduce(), etc.

```python
#Using lambda in map()
numbers = [1, 2, 3, 4, 5]
cubed = list(map(lambda x: x ** 3, numbers))
print(cubed)
```

```python
#Using lambda in filter()
numbers = [1, 8, 3, 10, 5, 13]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
#Using lambda in reduce()
from functools import reduce
numbers = [10, 12, 5, 2]
sum = reduce(lambda x, y: x + y, numbers)
print(sum)
```

**Q.1 Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.**

```python
from functools import reduce

def sum_of_evens(numbers):
    return reduce(lambda acc, x: acc + x if x % 2 == 0 else acc, numbers, 0)

numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_evens([1, 2, 3, 4, 5, 6]))
```

**Q.4 Write a Python function that checks if a given number is prime or not from 1 to 200.**

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter a number between 1 and 200: "))

if 1 <= num <= 200:
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
else:
    print("Please enter a number between 1 and 200.")
```

**Q.5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.**

```python
class Fibonacci:
    def __init__(self, num_terms):
        self.a, self.b, self.count, self.num_terms = 0, 1, 0, num_terms

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        self.a, self.b, self.count = self.b, self.a + self.b, self.count + 1
        return self.a if self.count > 1 else self.count - 1  # Return 0 for first term

num_terms = int(input("Enter the number of Fibonacci terms: "))
print(*Fibonacci(num_terms))
```

**Q.6  Write a generator function in Python that yields the powers of 2 up to a given exponent.**

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

max_exponent = int(input("Enter the maximum exponent: "))
for power in powers_of_two(max_exponent):
    print(power, end=" ")
```

**Q.10 Create a Python program that uses filter() to remove all the vowels from a given string.**




```python
def remove_vowels(string):
    vowels = "aeiouAEIOU"
    return "".join(filter(lambda character: character not in vowels, string))

input_string = input("Enter a string: ")
result = remove_vowels(input_string)
print("String without vowels:", result)
```
